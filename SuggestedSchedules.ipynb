```python
import json
import pandas as pd
```

```python
txt_directory = 'txt/'
json_directory = 'json/suggested/'
majors = ['act', 'ce','ee', 'cs']
addendum = '_suggested'
csv_ending = '.csv'
json_ending = '.json'
classes = dict()
```

```python
for major in majors:
    path = txt_directory + major + addendum + csv_ending
    df = pd.read_csv(path, header=0)
    for i in range(1, 5):
        df_year = df.where(df['year'] == i).dropna()
        classes[i] = dict()
        for s in ['F', 'S']:
            df_year_s = df_year.where(df['semester'] == s).dropna()
            classes[i][s] = []
            for index, row in df_year_s.iterrows():
                print(row)
                c = dict()
                c['section'] = row['section']
                c['number'] = row['number']
                classes[i][s].append(c)
    j_path = json_directory + major + addendum + json_ending
    with open(j_path, 'w') as jsonfile:
        json.dump(classes, jsonfile)
```
